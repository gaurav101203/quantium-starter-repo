```python
import pandas as pd
```

```python
df1 = pd.read_csv(".\data\daily_sales_data_0.csv")
df2 = pd.read_csv(".\data\daily_sales_data_1.csv")
df3 = pd.read_csv(".\data\daily_sales_data_2.csv")

combined = pd.concat([df1, df2, df3], ignore_index=True)

combined.to_csv(".\data\combined.csv", index=False)
```

```python
combined.info()
```

```python
combined_temp = combined.copy()
```

```python
combined_temp["price"] = combined_temp["price"].replace(r"\$", "", regex=True).astype(float)
```

```python
combined_temp["sales"] = combined_temp["price"] * combined_temp["quantity"]
```

```python
combined_temp
```

```python
combined_temp = combined_temp[combined_temp["product"] == "pink morsel"]
```

```python
combined_temp.drop("price", axis=1, inplace=True)
combined_temp.drop("product", axis=1, inplace=True)
combined_temp.drop("quantity", axis=1, inplace=True)
```

```python
combined_temp
```

```python
combined_temp.to_csv(".\data\combined.csv", index=False)
```
